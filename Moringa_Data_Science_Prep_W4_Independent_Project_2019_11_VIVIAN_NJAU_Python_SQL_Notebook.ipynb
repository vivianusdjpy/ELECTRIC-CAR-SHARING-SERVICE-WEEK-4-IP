# VALIDITY & ACCURACY

```
#drop the comments column since we will not use it for out analysis
#creating a copy
df_2 = df_1
df_2 = df_2.drop(['Displayed comment'], axis=1)
df_2.shape
```

```
#drop missing spaces
df_2['Address'] = df_2['Address'].str.strip()
df_2.tail(5)
```

```
#creating another copy
df_3 = df_2
df_3.shape
```

```
#drop the scheduled at column since  I will not use it in the analysis
df_3 = df_3.drop(['Scheduled at'], axis=1)
df_3.shape
```

# COMPLETENESS & UNIFORMITY

```
#creating a copy
df_4 = df_3
```

```
#Create a date column
df_4['Date'] = df_4["year"].map(str) +':'+ df_4["month"].map(str) +':'+ df_4["day"].map(str)
df_4.tail(5)
```

```
#create the time column
df_5 = df_4
df_5['Time'] = df_5["hour"].map(str) +':'+ df_5["minute"].map(str)
df_5
```

```
#create seconds column to convert the objects date and time to datetime format
#create a list of 5001 values with zeros
#assign the list to the column seconds
list_seconds = [0]*5000
df_5['seconds'] = list_seconds
df_5.info()
```

```
#drop the year, month, day, hour and minute columns
df_6 = df_5.drop(['year','month','day','hour','minute'],axis=1)
df_6
```

```
#creating the complete date time column
df_7 = df_6
df_7['Date_Time']= df_6["Date"].map(str) +':'+ df_6["Time"].map(str) +':'+df_6["seconds"].map(str)
df_7.tail(5)
```

```
#change the date data type to date format
from datetime import datetime
df_8 = df_7
df_8['Date_Time'] = pd.to_datetime(df_8['Date_Time'] , format='%Y:%m:%d:%H:%M:%S', errors = 'coerce')
df_8.info()
```

```
#drop date, time and second coulumns
df_9=df_8
df_9 = df_9.drop(['Date','Time','seconds',] , axis=1)
df_9.tail(5)
```
